```python
import requests
import io
import pandas as pd

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
response = requests.get(URL)
response.raise_for_status() # Raise an exception for bad status codes
dataset_part_2_csv = io.BytesIO(response.content)
df = pd.read_csv(dataset_part_2_csv)
df.head()
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the categorical plot
sns.catplot(
    x="FlightNumber",
    y="LaunchSite",
    hue="Class",
    data=df,
    kind="strip",   # or "swarm" for better visualization
    height=6,
    aspect=2
)

plt.title("Flight Number vs Launch Site (colored by class)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the categorical plot
sns.catplot(
    x="FlightNumber",
    y="LaunchSite",
    hue="Class",
    data=df,
    kind="strip",   # or "swarm" for better visualization
    height=6,
    aspect=2
)

plt.title("Flight Number vs Launch Site (colored by class)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    x="LaunchSite",
    y="PayloadMass",
    hue="Class",       # success/failure of mission
    data=df,
    kind="strip",      # can also try "swarm"
    height=6,
    aspect=2
)

plt.title("Payload Mass vs Launch Site")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(
    x="LaunchSite",
    y="PayloadMass",
    hue="Class",
    data=df
)

plt.title("Payload Mass Distribution by Launch Site")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate success rate by orbit
orbit_success = df.groupby('Orbit')['Class'].mean().reset_index()

# Plot success rate by orbit
plt.figure(figsize=(10,6))
sns.barplot(
    x='Orbit',
    y='Class',
    data=orbit_success,
    palette="viridis"
)

plt.title("Success Rate by Orbit Type")
plt.xlabel("Orbit Type")
plt.ylabel("Success Rate")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    x="FlightNumber",
    y="Orbit",
    hue="Class",       # shows success/failure
    data=df,
    kind="strip",      # or try "swarm" for better spacing
    height=6,
    aspect=2
)

plt.title("Flight Number vs Orbit Type (colored by success class)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.boxplot(
    x="Orbit",
    y="PayloadMass",
    hue="Class",     # success/failure
    data=df
)

plt.title("Payload Mass vs Orbit Type")
plt.xlabel("Orbit Type")
plt.ylabel("Payload Mass (kg)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure Date column is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract year
df['Year'] = df['Date'].dt.year

# Calculate success rate per year
yearly_success = df.groupby('Year')['Class'].mean().reset_index()

# Plot trend
plt.figure(figsize=(10,6))
sns.lineplot(
    x='Year',
    y='Class',
    data=yearly_success,
    marker='o'
)

plt.title("Yearly Launch Success Trend")
plt.xlabel("Year")
plt.ylabel("Success Rate")
plt.grid(True)
plt.show()
```
