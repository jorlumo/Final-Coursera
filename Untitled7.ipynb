```python
from math import radians, sin, cos, sqrt, atan2
import pandas as pd

# Haversine function
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in km
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c  # returns distance in km

# Example: Launch sites and proximities
launch_sites = {
    "CCAFS LC-40": [28.5623, -80.5774],
    "KSC LC-39A": [28.5733, -80.6469],
    "VAFB SLC-4E": [34.6328, -120.6108]
}

proximities = {
    "Nearest City": [28.3922, -80.6077],   # Cape Canaveral city
    "Nearest Railroad": [28.5630, -80.5670],
    "Nearest Highway": [28.5635, -80.5705]
}

# Calculate distances
distances = []
for site, (lat1, lon1) in launch_sites.items():
    for prox, (lat2, lon2) in proximities.items():
        dist_km = haversine(lat1, lon1, lat2, lon2)
        distances.append([site, prox, round(dist_km, 2)])

df_distances = pd.DataFrame(distances, columns=["Launch Site", "Proximity", "Distance (km)"])
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
